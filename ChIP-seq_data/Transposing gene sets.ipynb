```python
import pandas as pd
```

```python
file = open("NR3C1 ChIP-Seq.txt", "r")
lines = file.readlines()
lines
```

```python
for i in range(len(lines)):
    lines[i] = lines[i].split()

lines
```

```python
for i in range(len(lines)):
    lines[i] = pd.DataFrame(lines[i])
```

```python
for i in range(len(lines)):
    lines[i].to_csv('column{}.csv'.format(i + 1), index=False, header=False)
```
